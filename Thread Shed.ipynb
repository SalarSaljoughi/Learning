```python
# lets create separate lists for customers, sales, thread sold
```

```python
# looks like each transaction separated by , and the ;,;!
# in order to separate each transaction we need to first replace ;,; by something else and then separate by ,

daily_sales_replaced = daily_sales.replace(";,;", "_")
print(daily_sales_replaced)
```

```python
daily_transactions = daily_sales_replaced.split(",")
print(daily_transactions)
```

```python
# make a list of each individual transactions

daily_transactions_split = []
for transaction in daily_transactions:
    daily_transactions_split.append(transaction.split("_"))
print(daily_transactions_split)
```

```python
# looks like we have in consistant white spaces, lets remove them

transactions_clean = [] 
for transaction in daily_transactions_split:
    stripped_data = []
    for item in transaction:
        stripped_data.append(item.strip())
    transactions_clean.append(stripped_data)

print(transactions_clean)
```

```python
# now lets separat data into 3 list for customers, sales and thread sold

customers = []
sales = []
thread_sold = []

for item in transactions_clean:
        customers.append(item[0])
        sales.append(item[1])
        thread_sold.append(item[2])

print(customers)
print(sales)
print(thread_sold)
```

```python
# calculate total sales

total_sales = 0
for price in sales:
    sales_price = float(price.replace("$", ""))
    total_sales += sales_price
print(total_sales)
```

```python
# calculate how many of each color thread we sold today
# The end product we want here is a list that contains an item for each color thread sold, so no strings that have multiple colors in them.
print(thread_sold)
```

```python
thread_sold_split = []

for color in thread_sold:
    if "&" in color:
        separated = color.split("&")
        for ind_color in separated:
            thread_sold_split.append(ind_color)
    else:
        thread_sold_split.append(color)

print(thread_sold_split)
```

```python
# define function that counts given color

def color_count(color):
    number = thread_sold_split.count(color)
    print(number)
        
# test function
color_count('white')
        
```

```python
# text format message

colors = ["red", "yellow", "green", "white", "black", "blue", "purple"]

for diff_colors in colors:
    print(f"Thread Shed sold {color_count(diff_colors)} threads of {diff_colors} today.")
```
